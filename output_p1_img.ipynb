```python
import matplotlib.pyplot as plt
import os
import glob
from PIL import Image

plt.figure(figsize=(30,15))
fns = glob.glob(os.path.join('p1_output/*.png'))

print(len(fns))

for i,ele in enumerate(fns):
    if i > 31 : break
    ele = Image.open(ele)
    plt.subplot(4,8,i+1)
    plt.axis('off')
    plt.imshow(ele, interpolation='nearest')
    
plt.show()
```
